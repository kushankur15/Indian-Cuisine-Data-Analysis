```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
```

```python
filename = 'swiggy_vs_zomato_3000.csv'
df = pd.read_csv(filename)
df.head(1)
```

```python
df.isnull()
```

```python
df.columns
```

```python
df.describe()
```

```python
fig = px.scatter(x=df['swiggy_rating'], y=df['zomato_rating'])
fig.show()
```

```python
avg_cost = df.groupby('restaurant_type')['avg_cost_per_person_inr'].mean().reset_index()
fig = px.bar(
    avg_cost,
    x='restaurant_type',
    y='avg_cost_per_person_inr'
)
fig.show()
```

```python
summary = (
    df.groupby('restaurant_type')
    .agg({
        'avg_cost_per_person_inr':'mean',
        'swiggy_rating':'mean',
        'zomato_rating':'mean'
    })
    .reset_index()
)

long_df = pd.DataFrame({
    'restaurant_type': summary['restaurant_type'].tolist()*2,
    'avg_cost_per_person_inr': list(summary['avg_cost_per_person_inr']) * 2,
    'rating': list(summary['swiggy_rating']) + list(summary['zomato_rating']),
    'platform': ['Swiggy'] * len(summary) + ['Zomato'] * len(summary)
})
fig = px.bar(
    long_df,
    x='restaurant_type',
    y='avg_cost_per_person_inr',
    facet_row='platform',
    color='rating',
    color_continuous_scale='Viridis',
    range_color=[1, 5],
    title='Average Cost Per Person by Restaurant Type'
)
fig.show()
```

```python
df['cuisines'].value_counts().index
```

```python
cuisines = df['cuisines'].value_counts()
cuisines_list = cuisines.index
cuisines_counts = cuisines.values



for i,j in zip(cuisines_list,cuisines_counts):
    if j > 7:
        print(f"{i}: {j}")
    else:
        continue
```

```python
valid_cuisines = cuisines[cuisines>7].index
cuisine_df = df[df['cuisines'].isin(valid_cuisines)]
```

```python
avg_cost_cuisine = cuisine_df.groupby('cuisines')['avg_cost_per_person_inr'].median().reset_index()
avg_cost_cuisine
```

```python
fig = px.bar(
    avg_cost_cuisine,
    y='cuisines',
    x='avg_cost_per_person_inr',
    orientation='h'
)
fig.show()
```

```python
city_cost = df.groupby('city')['avg_cost_per_person_inr'].median().reset_index()
city_cost
```

```python
fig_1 = px.bar(
    city_cost,
    x='avg_cost_per_person_inr',
    y='city',
    orientation='h',
    color_discrete_sequence=['green']
)
fig_1.show()
```

```python
ratings_df = pd.melt(
    cuisine_df,
    id_vars='cuisines',
    value_vars=['swiggy_rating', 'zomato_rating'],
    var_name='platform',
    value_name='ratings'
)

ratings_df['platform'] = ratings_df['platform'].replace({
    'swiggy_rating': 'Swiggy',
    'zomato_rating': 'Zomato'
})
```

```python
ratings_df.head()
```

```python
ratings_df_new = ratings_df.groupby(['cuisines','platform'])['ratings'].mean().reset_index()
ratings_df_new
```

```python
ratings_df_new.groupby('platform')['ratings'].mean().reset_index()
```

```python
fig = px.bar(
    ratings_df_new,
    x='cuisines',
    y='ratings',
    color='platform',
    barmode='group'
)
fig.show()
```
